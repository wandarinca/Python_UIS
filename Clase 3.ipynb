````python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
data = pd.read_csv('iris.data.txt', names = ['sl', 'sw', 'pl', 'pw', 'label'])
````

````python
# Filtrar datos
data_setosa = data[data.label== 'Iris-setosa']
data_setosa.head()
````

````python
data2 = data[(data.label== 'Iris-setosa') & (data.pw > 0.4)] # se usa & en vez de and
data2
````

````python
data2.loc[:,'sw2'] = data2.sw**2
data2.head()
````

````python
import seaborn as sns
````

````python
pd.scatter_matrix(data, figsize=(12,12), diagonal='kde')
plt.show()
````

````python
#plt.scatter(data.sw, data.pw, )
colors = ['red', 'yellow', 'blue']
species = data.label.unique()
for i in range(len(colors)):
    tmp = data[data.label==species[i]]
    print(tmp.shape)
    plt.scatter(x=tmp.pl, y=tmp.pw, color=colors[i], label=species[i], alpha=.4)

plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()
````

````python
data.corr() # obtener la correlacion entre nuestras variables!
````

````python
plt.imshow(data.corr(), cmap=plt.cm.Reds, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(data.columns))]
plt.xticks(tick_marks, data.columns, rotation='vertical')
plt.yticks(tick_marks, data.columns)
plt.show()
````
